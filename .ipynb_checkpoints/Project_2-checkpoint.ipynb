```python
fields = ['Geography', 'Under 18 BelPov Est','Under 18 BelPov MOE' ]
df = pd.read_csv(filename,usecols=fields)
df.columns
```

```python
df.shape
```

```python
df.columns=["Name", "est","moe"]
df.head()
```

```python
df.tail(10)
```

```python
df['est']=df['est'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
df['moe']=df['moe'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
df.tail(10)
```

```python
df_valid=df.loc[df.est >= 0]
df_valid.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['zscore']=df['est']/(df['moe']/1.645) 
```

```python
df.loc[df.zscore < 1.645,['Name','est','moe']] # percent poverty = 0 (90% confidence)
```

```python
df.loc[df['Name']=='Buffalo city, New York']
```

```python
df['zBflo']=(df['est']-47.2)/((df['moe']/1.645)**2+(2.4/1.645)**2)**.5
df['zBflo'][:10]
```

```python
bufEst = df.loc[df.Name=='Buffalo city, New York',['est']]
bufMoe = df.loc[df.Name=='Buffalo city, New York',['moe']]
best = bufEst.values[0,0]
bmoe = bufMoe.values[0,0]
best,bmoe
```

```python
df['zBflo2']=(df['est']-best)/((df['moe']/1.645)**2+(bmoe/1.645)**2)**.5
df['zBflo2'][:10]
```

```python
df.loc[df.zBflo2 > 1.645,['Name','est','moe']] # percent poverty greater than Buffalo, 90% confidence
```

```python
df['Place'], df['State'] = df['Name'].str.split(', ', 1).str
df.head()
```

```python
overBflo= df.loc[df.zBflo2 > 1.645,['Place','State','est','moe']].reset_index()
overBflo.head()
```

```python
stateCount =overBflo.State.groupby(overBflo['State'])\
              .count()\
              .sort_values(ascending=False)
placeperst = stateCount.to_frame(name='counts').reset_index()
placeperst
```

### Based on the below data, Buffalo does not have the 4th highest poverty rate for children.
